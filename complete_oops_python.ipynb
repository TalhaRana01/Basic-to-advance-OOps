````python
class Car:
  # print("This is a car class")
  name = "Honda"
  model = "2020"
  
  def start(self):
    print("Car started")
  
  def stop(self):
    print("Car stopped")
  
  

car = Car()
print(car.name)
print(car.model)
car.start()  # This will give an error because start method does not have self parameter
car.stop()
print(type(car))
````
